.apply() method

```python
def last_four_of_cc(num):
    return str(num)[-4:]
```

```python
last_four_of_cc(123456789)
```

```python
df['CC Number'].apply(last_four_of_cc)
```

```python
df['last_four_of_cc'] = df['CC Number'].apply(last_four_of_cc)
df
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price >= 10 and price < 30:
        return '$$'
    else:
        return '$$$'
```

```python
yelp(32)
```

```python
df['yelp'] = df['total_bill'].apply(yelp)
df
```

```python
df[df['yelp'] == '$$$']
```

```python
def quality_of_tip(total_bill, tip):
    # if tip/total_bill > 0.25:
    #     return 'Generous'
    # else:
    #     return 'Other'

    return 'Generous' if tip/total_bill > 0.25 else 'Other'
```

```python
quality_of_tip(23.68, 11.84)
```

```python
df['quality_of_tip'] = df[['total_bill', 'tip']].apply(lambda df: quality_of_tip(df['total_bill'], df['tip']), axis=1)
```

```python
df[df['quality_of_tip'] == 'Generous']
```

```python
df['QualityOfTip'] = np.vectorize(quality_of_tip)(df['total_bill'], df['tip'])
```
